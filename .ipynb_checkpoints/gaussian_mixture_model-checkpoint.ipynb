## Gaussian Mixture Model
教師なし分類問題 clustering

### Basic Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Libraries for Machine Learning Model

```python
from sklearn import datasets
from sklearn import mixture, cluster
from sklearn import metrics, preprocessing
```

### 1. Data Preparation and Shaping Data

```python
data = datasets.load_wine()
df_wine = pd.DataFrame(data=data.data, columns=data.feature_names)
X = df_wine[['hue', 'proline']]
y = data.target

sc = preprocessing.StandardScaler() #平均0, 分散1に正規化
sc.fit(X)
X_norm = sc.transform(X)

plt.scatter(X_norm[:, 0], X_norm[:, 1], c=y)
```

### 4. Training Model and 5. Evaluating Model

```python
km=cluster.KMeans(n_clusters=3)
y_km=km.fit(X_norm)
plt.scatter(X_norm[:, 0],  X_norm[:, 1], c=y_km.labels_)
```

```python
gmm = mixture.GaussianMixture(n_components=3, covariance_type='full')
y_gmm = gmm.fit_predict(X_norm)
plt.scatter(X_norm[:, 0],  X_norm[:, 1], c=y_gmm)
```

### sklearn.mixture.GaussianMixture
#### parameters
- n_components: 線形結合するコンポーネント数
- covariance_type: 分散共分散行列
    - full
    - tied:
    - diag:
    - spherical
- tol:
- reg_covar: 正規化を行う
- max_iter: 最大繰り返し回数
