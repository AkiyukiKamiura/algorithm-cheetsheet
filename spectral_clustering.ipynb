## Spectral Clustering スペクトラルクラスタリング
教師なし分類問題 clustering

### Basic Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Libraries for Machine Learning

```python
from sklearn import datasets
from sklearn import cluster
from sklearn import metrics, preprocessing
```

### 1. Data Preparation and 3. Data Shaping

```python
X, y = datasets.make_moons(n_samples=200, noise=0.05, random_state=0)
sc = preprocessing.StandardScaler()
X_norm = sc.fit_transform(X)
plt.scatter(X_norm[:, 0], X_norm[:, 1], c=y)
```

### 4. Traning Model and 5. Evaluating Model

```python
km = cluster.KMeans(n_clusters=2)
km_y = km.fit(X_norm)
plt.scatter(X_norm[:, 0], X_norm[:, 1], c=km_y.labels_)
```

```python
spc = cluster.SpectralClustering(n_clusters=2, affinity="nearest_neighbors")
spc_y = spc.fit(X_norm)
plt.scatter(X_norm[:, 0], X_norm[:, 1], c=spc_y.labels_)
```
