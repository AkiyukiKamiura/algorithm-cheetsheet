### Basic Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Libraries for Learning Model

```python
from sklearn import datasets
from sklearn import cluster
from sklearn import metrics, preprocessing
```

### 1. Data Preparation

```python
data = datasets.load_wine()
df_wine = pd.DataFrame(data=data.data, columns=data.feature_names)
X = df_wine[['hue', 'proline']]
y = data.target

sc = preprocessing.StandardScaler() #平均0, 分散1に正規化
sc.fit(X)
X_norm = sc.transform(X)

plt.scatter(X_norm[:, 0], X_norm[:, 1], c=y)
```

### 4. Training Model

```python
km = cluster.KMeans(n_clusters=3)
km_y = km.fit(X_norm)
```

### 5. Evaluating Model

```python
plt.scatter(X_norm[:, 0], X_norm[:, 1], c=km_y.labels_)
plt.scatter(km_y.cluster_centers_[:,0],km_y.cluster_centers_[:,1],s=250, marker='x',c='red')
```
