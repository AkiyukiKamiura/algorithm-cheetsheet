## 損失関数一覧

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

### hinge loss:　ヒンジ関数
$$ L_{hinge}(x) = max(1-x, 0)$$

```python
def hinge(x):
    y = 1-x
    y[y<0] = 0
    return y
```

### squared hinge loss:　二乗ヒンジ関数
$$ L_{squared\_hinge}(x) = \{max(1-x, 0)\}^2$$

```python
def squared_hinge(x):
    y = 1-x
    y[y<0] = 0
    return np.power(y, 2)
```

### logistic loss:　ロジスティック関数
$$ L_{logistic}(x)　= \log(1 + \exp(-x) ) $$

```python
def logistic(x):
    return np.log(1 + np.exp(-x))
```

```python
x = np.arange(-2, 5, 0.1)
y_hinge = hinge(x)
y_shinge = squared_hinge(x)
y_log = logistic(x)
plt.plot(x, y_hinge, label='hinge')
plt.plot(x, y_shinge, label='squared hinge')
plt.plot(x, y_log, label='logistic')
plt.xlim(-2, 5)
plt.ylim(0, 3)
plt.legend()
```
